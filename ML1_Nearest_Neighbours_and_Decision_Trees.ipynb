# 1. Nearest Neighbor

basic hypothesis : samples which are close together in the feature space belong to the sample class

fit : indexing the position of all samples : c'est le training.

predict : compute euclidian distance to point in the training set. find the KNN (hyperparam). et prediction = majority of class among those. can be weighted by distance. par exemple on donne more wieght au plus proches.

hyper-parameters : Number of neighbours, distance metric (default euclidian but also manathan, ...), weights + preprocessing (PCA, normalization, ...)

The following example uses the Nearest Neighbor algorithm on the Histogram of Gradient decriptors in the dataset.

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=3)
clf.fit( dataset.train['hog'], dataset.train['labels'] )
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
# Predictions
pred = clf.predict(dataset.train['hog'])
# Accuracy Score
score = accuracy_score(dataset.train['labels'], pred)
print("Accuracy Score:", score)

# Confusion Matrix
cm = confusion_matrix(dataset.train['labels'], pred)
print("Confusion Matrix:\n", cm)
```

```python
from lab_tools import CIFAR10, get_hog_image

# Load Dataset
dataset = CIFAR10('CIFAR10/CIFAR10/')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# 1. Fraction of data for validation
p_val = 0.1
n_val = int(p_val * len(dataset.train['hog']))

# 2. Training and Validation Split
train_X = dataset.train['hog'][:-n_val]
train_Y = dataset.train['labels'][:-n_val]

val_X = dataset.train['hog'][-n_val:]
val_Y = dataset.train['labels'][-n_val:]

# KNN Classifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(train_X, train_Y)

# 3. Descriptive Score on Training Data
pred_descr = clf.predict(train_X)
descriptive_score = accuracy_score(train_Y, pred_descr)
print(f"Descriptive Score: {descriptive_score:.3f}")

# 4. Predictive Score on Validation
pred = clf.predict(val_X)
score = accuracy_score(val_Y, pred)
print(f"Predictive Score: {score:.3f}")

# 5. Confusion Matrix
cm = confusion_matrix(val_Y, pred)
print("Confusion Matrix:\n", cm)
print("Sum by Class:", cm.sum(axis=1))
```
